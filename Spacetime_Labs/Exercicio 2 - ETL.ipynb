```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandasql import sqldf
print ("Libraries sucessfully loaded") 
```

## ETL Routines

```python
# loading data_experiments file, hosted in Github repository for easy compatibility of this notebook when shared 
filename_exp = "https://raw.githubusercontent.com/tiagottmoraes/STL_test/main/Datasets/Dataset_case_DS.csv"
```

```python
df = pd.read_csv(filename_exp)
df
```

```python
df['data_colheita'] = pd.to_datetime(df['data_colheita']).dt.month
```

```python
df
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
df.rename(columns = {'data_colheita':'mes_colheita'}, inplace = True)
```

```python
df.rename(columns = {'bloco':'id_bloco'}, inplace = True)
```

```python
df.dtypes
```

### Checking NaN values present in the dataset

```python
print(" \nTotal number of NaN items present in the columns of the DataFrame 'Data': \n\n", df.isnull().sum())
```

```python
df=df.interpolate()
```

```python
df.isnull().sum()
```

#### Replacing the ramainder of NaN with the average of that variable

```python
avg_NDVI_b02 = df['NDVI_b02'].mean(axis=0)
avg_NDVI_b10 = df['NDVI_b10'].mean(axis=0)
avg_NDVI_b12 = df['NDVI_b12'].mean(axis=0)
```

```python
df['NDVI_b02'].replace(np.nan, avg_NDVI_b02, inplace=True)
df['NDVI_b10'].replace(np.nan, avg_NDVI_b10, inplace=True)
df['NDVI_b12'].replace(np.nan, avg_NDVI_b12, inplace=True)
```

```python
print(" \nTotal number of NaN items present in the columns of the DataFrame 'Data': \n\n", df.isnull().sum())
```

```python
df
```

```python
np.info((df['tch_bloco'].unique()))
```

```python
df1= sqldf(
"""
SELECT  id_bloco, count(talhao) as quant_talhoes, sum(area) as area_bloco, safra, mes_colheita, tch_bloco, AVG(NDVI_b01), AVG(NDVI_b02), 
        AVG(NDVI_b03), AVG(NDVI_b04), AVG(NDVI_b05), AVG(NDVI_b06), AVG(NDVI_b07),
        AVG(NDVI_b08), AVG(NDVI_b09), AVG(NDVI_b10), AVG(NDVI_b11), AVG(NDVI_b12)
FROM df
GROUP BY tch_bloco
ORDER BY safra, id_bloco
ASC
"""
)
df1
```

```python
df2= sqldf(
"""
SELECT  id_bloco, count(talhao) as quant_talhoes, sum(area) as area_bloco, safra, mes_colheita, tch_bloco, AVG(NDVI_b01)+ AVG(NDVI_b02)+ 
        AVG(NDVI_b03) as NDVI_1Q, AVG(NDVI_b04) + AVG(NDVI_b05) + AVG(NDVI_b06) as NDVI_2Q, AVG(NDVI_b07)+
        AVG(NDVI_b08) + AVG(NDVI_b09) as NDVI_3Q, AVG(NDVI_b10)+ AVG(NDVI_b11)+ AVG(NDVI_b12) as NDVI_4Q
FROM df
GROUP BY tch_bloco
ORDER BY safra, id_bloco
ASC
"""
)
df2
```

### Exporting the pre-processed df into a new .csv file
